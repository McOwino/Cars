## VISUALIZATION

```python
# importing visualizing libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cars1 = cars.drop(['MPG_City','MPG_Highway'], axis = 1)

cars1.head()
```

```python
# heatmap for correlation 
corr_matrix = cars1.corr()

# plot heatmap of correlation
plt.figure(figsize=(16,10))
sns.heatmap(corr_matrix, annot=True, cmap='Dark2')
```

### Conclusion from the heatmap
#### 1. There is a strong is strongly correlation between the  horsepower, invoice, MSRP, Engine size and cylinders.
#### 2. The weight and Cylinder are strongly correlated.
#### 3. There is also a strong correlation between the weight ,wheelbase and length.

```python
# Selecting the first five columns with categorical data
columns = cars.columns[2:5]

columns
```

```python
for column in columns:
    sns.countplot(x=cars[column], data=cars)
    plt.title(column)
    plt.figure(figsize=(16,10))
    sns.despine()
    plt.show()
```

#### With Europe having the biggest MSRP, what top 5 Makes and Models contribute to this result

```python
eu = (cars["Origin"] == 'Europe')

cars_europe = cars[eu]

cars_europe.head()
```

```python
cars_europe.shape
```

```python
# get value counts for brands
top_six = cars_europe.Make.value_counts()[:6]  
top_six
```

```python
# create another column called the newBrand
make_n = []

for make in cars_europe.Make:
    if make in top_six:
        make_n.append(make)
    else:
        make_n.append('other')
        
cars_europe['Make']=make_n  
cars_europe.Make.value_counts()
```

```python
value = cars_europe.Make.value_counts()

plt.pie(value, labels = value.index, shadow = True, startangle = 90)
plt.figure(figsize=(10,7))
plt.show() 
```

### What Model of Mercedes is on demand

```python
merce = (cars_europe['Make'] == 'Mercedes-Benz')

mercedez = cars_europe[merce]

mercedez
```

```python
sns.countplot(mercedez.Model, data=mercedez)
plt.title("Model")
plt.figure(figsize=(40,50))
sns.despine()
plt.show()
```
